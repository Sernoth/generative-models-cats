```python
import torch
from torchvision.utils import save_image

device = 'cuda'
noise_dim = 256

@torch.no_grad()
def generate(z):
    imgs = net_g(z)[0]          # FastGAN returns a list; [0] is the 256px image
    return (imgs + 1) / 2       # [-1,1] -> [0,1]

# 1. Sample a batch, pick two you like
torch.manual_seed(5)                      # fix seed so you can reproduce/save the z's
z_pool = torch.randn(16, noise_dim, device=device)
imgs_pool = generate(z_pool)
save_image(imgs_pool, "pool.png", nrow=4)  # look at this, choose 2 indices

# 2. Select the two latent vectors (change indices to your picks)
i, j = 4, 9
z0 = z_pool[i:i+1]
z1 = z_pool[j:j+1]

# 3. Linear interpolation: 10 points total (endpoints + 8 between)
alphas = torch.linspace(0, 1, 10, device=device).view(-1, 1)
z_interp = (1 - alphas) * z0 + alphas * z1   # shape [10, noise_dim]

# 4. Generate and present
imgs_interp = generate(z_interp)
save_image(imgs_interp, "interpolation3.png", nrow=10)

# Save the latents so the result is reproducible / documentable
torch.save({'z0': z0.cpu(), 'z1': z1.cpu(), 'z_interp': z_interp.cpu()},
           "interp_latents3.pt")
```

```python
# ============================================================
# Metrics against TWO real sets:
#   (A) full cat dataset  -> primary FID/KID/IS  (default subset ok)
#   (B) 100 training imgs -> memorization diagnostic (subset_size=100)
# Assumes eval/fake already populated.
# ============================================================
import os, torch
from torchvision import transforms
from PIL import Image
from cleanfid import fid
import torch_fidelity

im_size = 256

# >>> SET these two source folders:
FULL_SRC  = "cat_images"   # ~9k images
TRAIN_SRC = "cat_100"             # the exact 100 images you trained on

def prep_reals(src, dst):
    os.makedirs(dst, exist_ok=True)
    tf = transforms.Resize((im_size, im_size))   # match FastGAN training transform
    k = 0
    for fn in os.listdir(src):
        try:
            img = Image.open(os.path.join(src, fn)).convert("RGB")
        except Exception:
            continue
        tf(img).save(f"{dst}/{k:05d}.png")
        k += 1
    print(f"{dst}: {k} images")
    return k

n_full  = prep_reals(FULL_SRC,  "eval/real_full")
n_train = prep_reals(TRAIN_SRC, "eval/real_train")

def evaluate(real_dir, n_real, label):
    print(f"\n===== {label}  (reals: {n_real}) =====")
    kss = min(1000, n_real)

    fid_c = fid.compute_fid(real_dir, "eval/fake", mode="clean")
    # clean-fid uses num_subsets / subset_size (NOT kid_subsets / kid_subset_size)
    kid_c = fid.compute_kid(real_dir, "eval/fake", mode="clean",
                            num_subsets=100, subset_size=kss)
    print(f"clean-fid  FID: {fid_c:.3f}")
    print(f"clean-fid  KID: {kid_c:.5f}  (subset_size={kss})")

    # torch-fidelity DOES use kid_subset_size — that one was correct
    m = torch_fidelity.calculate_metrics(
        input1="eval/fake", input2=real_dir,
        isc=True, fid=True, kid=True, cuda=True, verbose=False,
        kid_subset_size=kss,
    )
    print(f"IS : {m['inception_score_mean']:.3f} ± {m['inception_score_std']:.3f}")
    print(f"FID(tf): {m['frechet_inception_distance']:.3f}")
    print(f"KID(tf): {m['kernel_inception_distance_mean']:.5f}")
    return fid_c, kid_c

fid_full,  kid_full  = evaluate("eval/real_full",  n_full,  "PRIMARY — full dataset")
fid_train, kid_train = evaluate("eval/real_train", n_train, "DIAGNOSTIC — 100 training imgs")

print("\n================ SUMMARY ================")
print(f"Primary    (full ):  FID {fid_full:.2f}   KID {kid_full:.5f}")
print(f"Diagnostic (train):  FID {fid_train:.2f}   KID {kid_train:.5f}")
print(f"Gap (train - full):  FID {fid_train - fid_full:+.2f}")
```

```python
def evaluate(real_dir, n_real, label):
    print(f"\n===== {label}  (reals: {n_real}) =====")
    kss = min(1000, n_real)   # KID subset can't exceed the smaller set

    # clean-fid: FID only (this version's compute_kid has no subset control)
    fid_c = fid.compute_fid(real_dir, "eval/fake", mode="clean")
    print(f"clean-fid  FID: {fid_c:.3f}")

    # torch-fidelity: KID + IS  (kid_subset_size works here)
    m = torch_fidelity.calculate_metrics(
        input1="eval/fake", input2=real_dir,
        isc=True, fid=True, kid=True, cuda=True, verbose=False,
        kid_subset_size=kss,
    )
    kid = m['kernel_inception_distance_mean']
    print(f"torch-fid  KID: {kid:.5f}  (subset_size={kss})")
    print(f"IS : {m['inception_score_mean']:.3f} ± {m['inception_score_std']:.3f}")
    print(f"FID(tf cross-check): {m['frechet_inception_distance']:.3f}")
    return fid_c, kid

fid_full,  kid_full  = evaluate("eval/real_full",  n_full,  "PRIMARY — full dataset")
fid_train, kid_train = evaluate("eval/real_train", n_train, "DIAGNOSTIC — 100 training imgs")

print("\n================ SUMMARY ================")
print(f"Primary    (full ):  FID {fid_full:.2f}   KID {kid_full:.5f}")
print(f"Diagnostic (train):  FID {fid_train:.2f}   KID {kid_train:.5f}")
print(f"Gap (train - full):  FID {fid_train - fid_full:+.2f}")
```
